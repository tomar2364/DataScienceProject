## 1>The dataset is highly skewed toward the cities included in Delhi-NCR. So, we will summarise all the other cities in Rest of India while those in New Delhi, Ghaziabad, Noida, Gurgaon, Faridabad to Delhi-NCR. Doing this would make our analysis turn toward Delhi-NCR v Rest of India.
### 1:Plot the bar graph of number of restaurants present in Delhi NCR vs Rest of India.

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("zomato.csv",encoding="ISO-8859-1")
data=df[df["Country Code"]==1]
data1=data[(data["City"]=="New Delhi")|(data["City"]=="Ghaziabad")|(data["City"]=="Noida")|(data["City"]=="Gurgaon")|(data["City"]=="Faridabad")]
count=data1["City"].value_counts()
count
data2=data[(data["City"]!="New Delhi")&(data["City"]!="Ghaziabad")&(data["City"]!="Noida")&(data["City"]!="Gurgaon")&(data["City"]!="Faridabad")]
count2=data2["City"].value_counts()
ncr=dict(count)
other=dict(count2)
key1=list(ncr.keys())
value1=list(ncr.values())
key2=list(other.keys())[0:18]
value2=list(other.values())[0:18]
key3=list(other.keys())[18:]
value3=list(other.values())[18:]
print("City(Delhi NCR)\t |  Number of Restaurants")
for i in range(len(key1)):
    print('{:<17s}|{:>17d}'.format(key1[i],value1[i]))
plt.bar(key1,value1,width=0.3)
plt.xlabel("CITY (DELHI NCR)")
plt.ylabel("No. OF RESTAURANTS")
plt.show()
print("City(Other)\t |  Number of Restaurants")
for i in range(len(key2)):
    print('{:<17s}|{:>17d}'.format(key2[i],value2[i]))
plt.bar(key2,value2,width=0.5)
plt.xlabel("CITY (REST OF INDIA)")
plt.ylabel("No. OF RESTAURANTS")
plt.xticks(rotation=80)
plt.show()
print("City(Other)\t |  Number of Restaurants")
for i in range(len(key3)):
    print('{:<17s}|{:>17d}'.format(key3[i],value3[i]))
plt.bar(key3,value3,width=0.5)
plt.xlabel("CITY (REST OF INDIA)")
plt.ylabel("No. OF RESTAURANTS")
plt.xticks(rotation=60)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv("zomato.csv",encoding="ISO-8859-1")
data=df[df["Country Code"]==1]
xpos = np.arange(1,11000,1000)
dataUsed = data
dataUsed = dataUsed[dataUsed['Country Code'] == 1]
Means = []
for i in range(1,11000,1000):
    dataUsed1 = dataUsed[(dataUsed['Votes'] >= i) & (dataUsed['Votes'] < i+1000)]
    if(len(dataUsed1) == 0):
        mean = 0                       ## Appending 0 else nan value will be appended
    else:
        mean = dataUsed1['Aggregate rating'].mean()
    
    Means.append(mean)
    
    
for i in range(0,len(Means)-1,1):
    if(Means[i] == 0):
        Means[i] = (Means[i-1] + Means[i+1])/2
        ## Filling the average if mean == 0 i.e. data was empty
    
dg=[]    
for i in range(len(Means)):
    print("Mean of No. of Votes between",1000*i,'to',1000*i+999, "is :",Means[i])
for i in Means:
    dg.append(round(i,2))
dh=["0-999","1000-1999","2000-2999","3000-3999","4000-4999","5000-5999","6000-6999","7000-7999","8000-8999","9000-9999","10000-10999"]

plt.plot(dh,dg,color='blue',linewidth=3)
plt.xticks(rotation=60)
plt.grid()
plt.xlabel("Number Of Votes")
plt.ylabel("Mean Of Aggregate Rating")
plt.show()
```

###  2: Plot the histogram of aggregate rating of restaurant( drop the unrated restaurant).

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv("zomato.csv",encoding="ISO-8859-1")
dataUsed=df
dataUsed = dataUsed[dataUsed['Country Code'] == 1]
dataUsed["Aggregate rating"].replace(0,np.nan,inplace=True)
df=dataUsed["Aggregate rating"].dropna()
d1=list(df)
d={}
for i in d1:
    d[i]=d.get(i,0)+1
d
sum1=0
sum2=0
sum3=0
sum4=0
for i,j in d.items():
    if i>=0 and i<2:
        sum1=sum1+j
    elif i>=2 and i<3:
        sum2=sum2+j
    elif i>=3 and i<4:
        sum3=sum3+j
    elif i>=4 and i<=5:
        sum4=sum4+j
res=[sum1,sum2,sum3,sum4]
data=["0-1.9","2-2.9","3-3.9","4-5"]
print("Aggregate Rating\t|        Frequecy Of Restaurants")
for i in range(len(res)):
    print('{:<24s}|{:>24d}'.format(data[i],res[i]))
xpos = np.arange(1,5.2,0.2)
plt.hist(res,bins = xpos,edgecolor = 'black',width = 0.2,)
plt.show()
```
